# 02. PyTorch Neural Network Classification

**What is a Classification problem?**

- Predicting whether something is one thing or other i.e. An image is  of Dog or Cat?, email is spam or not spam?

**Classification Problem Type:**
1. **Binary Classification:**
    - One of two options: (yes / no)
    - eg. Predict whether or not someone has heart disease based on their health parameters.

2. **Multi-class Classification:**
    - more than one or two options
    - eg. Decide whether a photo is of food, a person or a dog.

3. **Multi-label Classification:**
    - Target can be assigned more than one option
    - eg. Predict what categories should be assigned to a Wikipedia article (e.g. mathematics, science & philosophy).

Except instead of trying to predict a straight line (predicting a number, also called a regression problem), we'll be working on a classification problem.

## 3.Building a PyTorch Classification Model

Building a model to classify the red and blue dots

To do so, we want to:
1. setup device agnostic code so our code will run on an accelerator (GPU) if there is one
2. Constrcut a model (by sub classing `nn.Module`)
3. define a loss function and optimizer
4. Create a training and test loop


```python
# import PyTorch and nn
import torch
import torch.nn as nn

# Make device agnoctic code
device = "cuda" if torch.cuda.is_available else "cpu"
device # it is alreaady running on the GPU i.e. cuda
```

### Classification Model Implementation

1. subclass `nn.module` almost every model in PyTorch do this
2. Create 2 `nn.Linear()` layers that are capable of handling the shapes of our data
3. Define a `forward()` method
4. instatiate an instance of our model class and send it to the target `device`


```python
# 1. Construct a model that subclasses `nn.Module`
class CircleModelClassification(nn.Module):
  def __init__(self):
    super().__init__()
    # 2. Create 2 nn.Linear layers to handel the shapes of our data
    self.layer_1 = nn.Linear(in_features = 2, out_features = 5) # Takes in 2 features and upscales it to 5
    # out_features of previous layer must match the in_features of next layer
    self.layer_2 = nn.Linear(in_features = 5, out_features = 1)
    # out_features of the last layer must be the shape of `Y`

  # 3. Defining a forward method that outline a forward pass
  def forward(self, X):
    return self.layer_2(self.layer_1(X))
    # X (goes into) --> layer_1 (output of layer_1, goes into) --> layer_2 --> output
```

```python
# 4. Instantiate an instance of our model class and send it to target device
model_0 = CircleModelClassification()
model_0
```

### Setup Loss function and Optimizer

1. `nn.L1Loss` : Loss Function
2. `torch.optim.SGD()` : SGD Optimizer

For example,

**For regression** we want MAE or MSE (mean absolute error or mean squared error)

**For Classification** we want `BCELoss` binary cross entropy loss or categorical cross entropy (cross entropy), It's also known as logarithmic loss or log loss.

### Loss Function fot Classsification Problem:

1. **Binary Cross Entropy (`nn.BCELoss`)** is the negative average of the log of corrected predicted probabilities.

- Input Requirements:

    Expects the input predictions (y_pred) to be probabilities (values between 0 and 1), typically obtained by applying a Sigmoid activation function on raw logits beforehand.

- Target values (y_true) must also be in the range [0, 1].

---

2. **Binary Cross Entropy with Logits Loss:** `nn.BCEWithLogitsLoss()`
  - `nn.BCEWithLogitsLoss()` has sigmoid activation function built in
  - BCEWithLogitsLoss
    Input Requirements:

    Expects raw logits (unbounded real values) as input predictions (y_pred).

  - The target values (y_true) are in the range [0, 1].
    Built-in Sigmoid:

  - Automatically applies the Sigmoid activation function internally on the logits before computing the binary cross-entropy loss.

  - This makes it numerically more stable because it avoids redundant Sigmoid computations in the forward and backward passes.

  ---

Conclusion:
1. nn.BCELoss(y_preds, y_train)
2. nn.BCEWithLogitsLoss(y_logits, y_train)

### Optimizer for Classification Problem

- For classification problem two famous Optimizers are `torch.optim.SGD` , `torch.optim.Adam`

```python
# Loss function and optimizer
loss_fn = nn.BCEWithLogitsLoss() # BCEWithLogitsLoss = sigmoid activation function
optimizer = torch.optim.SGD(model_0.parameters(), lr = 0.01)
```

### Calculating Accuracy of our model

- Accuracy: out of 100 sample what percentage does our model get the right predictions?


```python
def accuracy_fn(y_true : torch.tensor, y_preds : torch.tensor) -> float:
  correct = torch.eq(y_true, y_preds).sum().item()
  return (correct / len(y_preds)) * 100
```

## Concept of Logits in Machine Learning

- **Logits** : Our model's output are going to be raw **Logits**

- We can convert this raw *Logits* into predictions by passing them into some sort of Activation Functions (e.g. sigmoid for binary cross entropy, and softmax for multi-class classification).

- Then we can convert our model's predictions probabilities to **prediction labels** by either roounding them or taking the `argmax()` for multi class classification

##4.Training a Model

-  Making Training and Testing loop

Training Loop:

1. Set model to train mode
2. Forward pass
3. Training Loss calculation
4. Optimizer zero grad
5. Loss backward (backpropogation)
6. Optimizer step (grdient descent, update the parameters)

---

Testing Loop:
1. Set model to evaluating mode
2. with torch.inference_mode()
3. Forward pass
4. Testing Loss calculation

## BCELoss and BCEWithLogitsLoss

1. nn.BCELoss(torch.sigmoid(X_train), X_train)
2. nn.BCEWithLogitsLoss(Y_Logits, X_train)

```python
epochs = 1000
for epoch in range(epochs):

  # Training Loop
  model_0.train()
  # forward pass
  Y_logits = model_0(X_train).squeeze()
  Y_preds = torch.round(torch.sigmoid(Y_logits))
  # Loss Calculation and Accuracy of our Model
  loss_training = loss_fn(Y_logits, # nn.BCEWithLogitsLoss expects raw Logits
                          Y_train)
  train_accuracy = accuracy_fn(Y_train, Y_preds)
  # Zero grad
  optimizer.zero_grad()
  # Loss Backward
  loss_training.backward()
  # Update the Parameters (weights and bias)
  optimizer.step()

  # Testing Loop
  model_0.eval()
  with torch.inference_mode():
    # Forward pass:
    y_logits_test = model_0(X_test).squeeze()
    y_preds_test = torch.round(torch.sigmoid(y_logits_test))
    # Loss Calculation:
    loss_testing = loss_fn(y_logits_test, Y_test)
    test_accuracy = accuracy_fn(Y_test, y_preds_test)

    if epoch % 500 == 0:
      print(f"Epoch: {epoch} | Training Loss: {loss_training:.5f} | Training Acc: {train_accuracy:.2f}%| Test Acc: {test_accuracy:.2f}% | Testing Loss: {loss_testing:.5f}")
```

### **What we can do to Improve Our Model?**

- Add more hidden units (neurons)
- Adding the Activation Function (Non - Linearity)
- Changing the learning rate
- Change the loss Function
- Change the optimization Function

### Updating Layers and Activation Functiuon

```python
class MyUpdatedModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(in_features=2, out_features=512),
        nn.ReLU(), # Non - Linear Activation after Layer 1
        nn.Linear(in_features=512, out_features=512),
        nn.ReLU(), # Non - Linear Activation after Layer 2
        nn.Linear(in_features=512, out_features=512),
        nn.ReLU(),# Non - Linear Activation after Layer 3
        nn.Linear(in_features=512, out_features=1),
    )
  def forward(self, X : torch.tensor) -> torch.tensor :
    return self.model(X)
```

##7.What is Non - Linearity?

- A relation which is not Linear!

## Non - Linear Activation Functions?
- ReLU()
- Tanh()
- Sigmoid()

### Replicating Non-Linearity using Python

####1.ReLU()

```python
X_linear = torch.arange(-10, 10, 0.1)
X_linear.shape
```

#### Plotting this Linear Curve

```python
plt.plot(X_linear)
```

#### Transforming this into Non Linear Function

```python
def relu(X: torch.tensor) -> torch.tensor:
  for i in range(len(X)):
    if X[i] > 0:
      pass
    elif X[i] < 0:
      X [i] = 0
  return X
```

```python
## Make it One Linear
def one_linear_relu(X: torch.tensor) -> torch.tensor:
  return torch.where(X<0,0,X)
```

```python
## Another Method of Doing the same is,
def ReLU(X: torch.tensor) -> torch.tensor:
  return torch.max(X, torch.tensor(0, dtype = X.dtype))
```

```python
X_Non_Linear = one_linear_relu(X_linear)
plt.plot(X_Non_Linear)
```

```python
X_Non_Linear = relu(X_linear)
plt.plot(X_Non_Linear)
```

```python
X_Non_Linear = ReLU(X_linear)
plt.plot(X_Non_Linear)
```

####2.Sigmoid()

```python
def sigmoid(X: torch.tensor) -> torch.tensor:
  return 1/(1 + torch.exp(-X))
```

```python
X_Non_Linear = sigmoid(X_linear)
plt.plot(X_Non_Linear)
```

####3.Tanh()

```python
def Tanh(X: torch.tensor) -> torch.tensor:
  exp_X = torch.exp(X)
  nev_exp_X = torch.exp(-X)
  return ((exp_X - nev_exp_X)/(exp_X + nev_exp_X))
```

```python
X_Non_Linear = Tanh(X_linear)
plt.plot(X_Non_Linear)
```

#### Creating a Toy data set for Multi class classsification

```python
import torch
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

# set hyperparamter for our model

## From sklearn.datasets.make_blobs():

# 1. Create multi-class data
NUM_CLASSES = 4
NUM_FEATURES = 2
RANDOM_SEED = 43
X_blob, Y_blob = make_blobs(n_samples = 1000,
                     n_features = NUM_FEATURES,
                     centers = NUM_CLASSES,
                     cluster_std =  1.7,
                     random_state = RANDOM_SEED)

# 2. Turn this data into tensors
X_blob = torch.from_numpy(X_blob).type(torch.float)
Y_blob = torch.from_numpy(Y_blob).type(torch.LongTensor) # Because `CrossEntropyLoss()` takes long as target values

# 3. Split into training and test data sets
X_blob_train, X_blob_test, Y_blob_train, Y_blob_test = train_test_split(X_blob, Y_blob, test_size=0.2)

# 4. Plot the data
plt.figure(figsize =(10, 7))
plt.scatter(X_blob[:, 0], X_blob[:, 1], c = Y_blob, cmap = plt.cm.RdYlBu)
```

#### Building a multi-class classsification Model

```python
# # Create device agnostic code
# device = "cuda" if torch.cuda.is_available() else "CPU"
# device
```

```python
# Build multi class classifiaction Model
import torch
import torch.nn as nn
class MyMultiClassClassification(nn.Module):
  def __init__(self, in_features, out_features, hidden_units = 8):
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(in_features, hidden_units),
        # nn.ReLU(),
        nn.Linear(hidden_units, hidden_units),
        # nn.ReLU(),
        nn.Linear(hidden_units, hidden_units),
        # nn.ReLU(),
        nn.Linear(hidden_units, hidden_units),
        # nn.ReLU(),
        nn.Linear(hidden_units, out_features)
    )
  def forward(self, X: torch.tensor) -> torch.tensor :
    return self.model(X)
# Create an instance of our Model and send it to target device
torch.manual_seed(42)
model_3 = MyMultiClassClassification(2, 4, 8)
model_3
```

#### Loss Function and optimizer

```python
loss_fn_multi = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model_3.parameters(), lr = 0.01)
```

#### Getting Prediction Probabilities for a multi class classification

## **What we'll do is:**

logits (raw output of model class)-> preds prob (Applying softmax activation) -> preds labels (argmax(preds_prob))

```python
# X_blob_train, X_blob_test = X_blob_train.to(device), X_blob_test.to(device)
# Y_blob_train, Y_blob_test = Y_blob_train.to(device), Y_blob_test.to(device)
```

```python
model_3.eval()
with torch.inference_mode():
  y_logits = model_3(X_blob_test)
y_preds_prob = torch.softmax(y_logits, dim = 1)
y_preds_prob[:10]
```

#### Building a training Loop for our multi-class classsification model

```python
epochs = 5000
# X_blob_train, X_blob_test = X_blob_train.to(device), X_blob_test.to(device)
# Y_blob_train, Y_blob_test = Y_blob_train.to(device), Y_blob_test.to(device)

for epoch in range(epochs):
  # Training
  model_3.train()

  y_train_logits = model_3(X_blob_train).type(torch.float)
  y_train_preds_labels = torch.softmax(y_train_logits, dim = 1).argmax(dim = 1)


  loss_train = loss_fn_multi(y_train_logits, Y_blob_train)

  acc_training = accuracy_fn(Y_blob_train, y_train_preds_labels)

  optimizer.zero_grad()

  loss_train.backward()

  optimizer.step()

# Testing
  model_3.eval()
  with torch.inference_mode():
    y_test_logits = model_3(X_blob_test)
    y_test_preds_labels = torch.softmax(y_test_logits, dim = 1).argmax(dim = 1)

    loss_testing = loss_fn_multi(y_test_logits, Y_blob_test)

    acc_testing = accuracy_fn(Y_blob_test, y_test_preds_labels)

    if epoch % 100 == 0:
      print(f"Epoch {epoch} | Train Acc: {acc_training:.2f}% | Trainign Loss {loss_train:.5f}% |  Train Acc: {acc_testing:.2f}% | Testing Loss {loss_testing:.5f}%")
```
